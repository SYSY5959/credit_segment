```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import f1_score
import pandas as pd
import numpy as np

# ✅ 1. 데이터 로딩 (df: train, test: test set)
# df와 test는 이미 pandas DataFrame으로 존재한다고 가정

# ✅ 2. 학습 데이터 분리
# X, y 분리
X = df.drop(columns=['Segment_A', 'Segment_B', 'Segment_C', 'Segment_D', 'Segment_E']).values
y_onehot = df[['Segment_A', 'Segment_B', 'Segment_C', 'Segment_D', 'Segment_E']].values
y = np.argmax(y_onehot, axis=1)

# ✅ 3. Tensor 변환
# CNN을 위해 (batch, channels=1, length=features) 형태로 reshape
X_tensor = torch.tensor(X, dtype=torch.float32).unsqueeze(1)  # (N, 1, 947)
y_tensor = torch.tensor(y, dtype=torch.long)

# ✅ 4. DataLoader 정의
train_dataset = torch.utils.data.TensorDataset(X_tensor, y_tensor)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=128, shuffle=True)

# ✅ 5. 1D CNN 모델 정의
class CNN1DClassifier(nn.Module):
    def __init__(self, input_length, output_dim=5):
        super(CNN1DClassifier, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv1d(in_channels=1, out_channels=32, kernel_size=5, padding=2),
            nn.BatchNorm1d(32),
            nn.ReLU(),
            nn.MaxPool1d(2),  # (N, 32, input_length/2)

            nn.Conv1d(32, 64, kernel_size=5, padding=2),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.MaxPool1d(2),  # (N, 64, input_length/4)
        )

        conv_out_dim = input_length // 4 * 64  # Flatten 후 차원 계산
        self.fc = nn.Sequential(
            nn.Linear(conv_out_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, output_dim)
        )

    def forward(self, x):
        x = self.cnn(x)
        x = x.view(x.size(0), -1)  # Flatten
        return self.fc(x)

# ✅ 6. 모델 초기화
model = CNN1DClassifier(input_length=X.shape[1])
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

# ✅ 7. 학습 루프
for epoch in range(10):
    model.train()
    all_preds = []
    all_labels = []

    for xb, yb in train_loader:
        preds = model(xb)
        loss = criterion(preds, yb)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        predicted = torch.argmax(preds, dim=1)
        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(yb.cpu().numpy())

    f1 = f1_score(all_labels, all_preds, average='macro')
    print(f"Epoch {epoch+1}: F1 Macro (Train) = {f1:.4f}")

# ✅ 8. 테스트 데이터 예측
X_test = test.values
X_test_tensor = torch.tensor(X_test, dtype=torch.float32).unsqueeze(1)  # (N, 1, 947)

model.eval()
with torch.no_grad():
    test_preds = model(X_test_tensor)
    test_pred_classes = torch.argmax(test_preds, dim=1).cpu().numpy()

# ✅ 9. 예측 결과를 원핫 인코딩으로 변환
test_pred_onehot = np.eye(5)[test_pred_classes]

# ✅ 10. 결과 저장
result_df = pd.DataFrame(test_pred_onehot, columns=['Segment_A', 'Segment_B', 'Segment_C', 'Segment_D', 'Segment_E'])
result_df.to_csv("submission.csv", index=False)
print("submission.csv 저장 완료!")
```

# TabNet

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class GLU_Block(nn.Module):
    def __init__(self, in_features, out_features):
        super().__init__()
        self.fc = nn.Linear(in_features, out_features * 2)
        self.bn = nn.BatchNorm1d(out_features * 2)

    def forward(self, x):
        x = self.fc(x)
        x = self.bn(x)
        x_a, x_b = x.chunk(2, dim=-1)
        return x_a * torch.sigmoid(x_b)

class FeatureTransformer(nn.Module):
    def __init__(self, dim, shared_layers=None):
        super().__init__()
        self.shared = shared_layers
        self.step_dependent = nn.Sequential(
            GLU_Block(dim, dim),
            GLU_Block(dim, dim)
        )

    def forward(self, x):
        if self.shared is not None:
            x = self.shared(x)
        return self.step_dependent(x)

class AttentiveTransformer(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.fc = nn.Linear(dim, dim)
        self.bn = nn.BatchNorm1d(dim)

    def forward(self, x, prior):
        x = self.fc(x)
        x = self.bn(x)
        x = x * prior
        return F.sparsemax(x, dim=-1)  # sparsemax 대신 softmax도 가능

class TabNet(nn.Module):
    def __init__(self, input_dim, output_dim, n_d=64, n_a=64, n_steps=5, gamma=1.5):
        super().__init__()
        self.n_d = n_d
        self.n_a = n_a
        self.n_steps = n_steps
        self.gamma = gamma

        self.initial_bn = nn.BatchNorm1d(input_dim)

        shared = nn.Sequential(
            GLU_Block(input_dim, n_d + n_a),
            GLU_Block(n_d + n_a, n_d + n_a)
        )

        self.steps = nn.ModuleList()
        for _ in range(n_steps):
            transformer = FeatureTransformer(n_d + n_a, shared_layers=shared)
            attention = AttentiveTransformer(n_a)
            self.steps.append(nn.ModuleDict({
                'transformer': transformer,
                'attention': attention
            }))

        self.final_fc = nn.Linear(n_d, output_dim)

    def forward(self, x):
        x = self.initial_bn(x)
        B, D = x.shape
        prior = torch.ones(B, D, device=x.device)
        output = torch.zeros(B, self.n_d, device=x.device)
        masked = x

        for step in self.steps:
            x_transformed = step['transformer'](masked)
            d = x_transformed[:, :self.n_d]
            a = x_transformed[:, self.n_d:]
            output += F.relu(d)
            mask = step['attention'](a, prior)
            masked = mask * x
            prior = prior * (self.gamma - mask)

        return self.final_fc(output)
```
