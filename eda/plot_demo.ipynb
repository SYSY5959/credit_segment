```python
import pandas as pd
import numpy as np
import random

def create_test_df(seed):

    random.seed(seed)
    np.random.seed(seed)

    # Segment 구성
    segments = ['A'] * 30 + ['B'] * 20 + ['C'] * 20 + ['D'] * 15 + ['E'] * 15

    # 연속형 컬럼: 0~100 사이 랜덤 float
    continuous_vals = np.random.uniform(0, 100, size=100)

    # 범주형 컬럼: 'X', 'Y', 'Z' 중 하나
    categorical_vals = random.choices(['X', 'Y', 'Z'], k=100)

    df = pd.DataFrame({
        'ID': [f'user_{i+1}' for i in range(100)],
        'Segment': segments,
        '연속값': continuous_vals,
        '범주값': categorical_vals
    })
    
    return df

dfs = [create_test_df(i) for i in range(6)]
```
